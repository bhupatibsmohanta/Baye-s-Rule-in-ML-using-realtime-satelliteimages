# Step 3

```python
#Calculating covarianc_matrix of River_Class
cov_riv=np.cov([img1[riv1[:,0],riv1[:,1],0],img2[riv2[:,0],riv2[:,1],0],img3[riv3[:,0],riv3[:,1],0],img4[riv4[:,0],riv4[:,1],0]])
cov_riv
```

# Step 4

```python
#Calculating covarianc_matrix of NonRiver_Class
cov_non=np.cov([img1[non1[:,0],non1[:,1],0],img2[non2[:,0],non2[:,1],0],img3[non3[:,0],non3[:,1],0],img4[non4[:,0],non4[:,1],0]])
cov_non
```

# Step 7 - 14

Given, prior probabilities P1=0.3 and P2=0.7

```python
output=[]
P1 = 0.3
P2 = 0.7

for i in test_data:
    #Step 8
    x=np.array(i-T1)
    inverse=np.linalg.inv(cov_riv)
    river_class = float(np.dot(np.dot(x.reshape(1,4), inverse), x))
    
    #Step 9
    y=np.array(i-T2)
    inverse=np.linalg.inv(cov_non)
    non_class = float(np.dot(np.dot(y.reshape(1,4), inverse), y))
    
    #Step 10 - calculate density funtion for River_Class
    p1= (0.5)*(1/np.sqrt(np.linalg.det(cov_riv)))*np.exp((-0.5)*river_class)
    #Step 11 - calculate density funtion for NonRiver_Class
    p2= (0.5)*(1/np.sqrt(np.linalg.det(cov_non)))*np.exp((-0.5)*non_class)
    
    #Step 11 Applying Baye' Rule and stroring in output_image
    if(P1*p1 >= P2*p2):
        output.append(255)
    else:
        output.append(0)
        
    
output
```

```python
#reshape the image to required scale (512*512)
output=np.array(output).reshape(512,512)
output
```

```python
#showing the output_image
import numpy as np
from PIL import Image

img = Image.fromarray(np.uint8(output) , 'L')
img.show()              ##Look into pop up window.
```
