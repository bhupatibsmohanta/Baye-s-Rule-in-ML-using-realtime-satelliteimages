```python
import cv2
img1=cv2.imread('1.jpg')
img2=cv2.imread('2.jpg')
img3=cv2.imread('3.jpg')
img4=cv2.imread('4.jpg')
```

```python
img4.shape
```

```python
riv1=[]

def mousePoints(event,x,y,flags,params):
    if(event==cv2.EVENT_LBUTTONDOWN):
        print(img4[y,x])                       #pixel values corresponding to the given coordinates.
        riv1.append([y,x])
        

cv2.imshow("4_Riv",img4)
cv2.setMouseCallback("4_Riv",mousePoints)
cv2.waitKey(0)

## Click 50 time inside the River to collect the coordinates of river class
```

```python
import numpy as np
riv1=np.array(riv1)
print(riv1.shape)
riv1            ##Copy these coordinates into Assignment1 program for river class.
```
